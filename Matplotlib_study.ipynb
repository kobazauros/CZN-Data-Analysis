## Task 3

Using 'np.linspace' create 'x' array of 51 numbers from -3 to 3 including both numbers.  
Create arrays $y_1$, $y_2$, $y_3$, $y_4$ based on the following formulas:  
$$y_1 = x^2$$
$$y_2 = 2 * x + 0.5$$
$$y_3 = -3 * x - 1.5$$
$$y_4 = \sin(x)$$
Using 'subplots' function from 'matplotlib.pyplot' module, create 'matplotlib.figure.Figure' named 'fig' and axis array named 'ax'. Prepare 4 different graphs in the grid of 2 rows and 2 colums with 'x' used as horizontal axis.

For upper left graph use $y_1$, for в upper right graph use - $y_2$, for lower left use - $y_3$, for lower right use  - $y_4$ as a vertical axis.  
Make titles for graphs: '$y_1$ graph', '$y_2$ grapf' etc.  
For the graph in lower left corner set limits for 'x' axis with -5 до 5.  
Set the size of 'fig' figure as 8 for width and 6 hor height.  
The space between graphs should be 0.3.

```python
x = np.linspace(-3,3,51)
y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)
fig, ax = plt.subplots(2, 2, gridspec_kw={'wspace':0.3, 'hspace':0.3}, figsize=(8, 6))
ax[0, 0].plot(x, y1)
ax[0, 0].set_title('y1 graph')
ax[0, 0].set_xlim(-5,5)
ax[0, 1].plot(x, y2)
ax[0, 1].set_title('y2 graph')
ax[1, 0].plot(x, y3)
ax[1, 0].set_title('y3 graph')
ax[1, 1].plot(x, y4)
ax[1, 1].set_title('y4 graph')
```


## Task 4

For this task we shall use 'creditcard.csv' dataset containing credit card fraud metrics:
Credit Card Fraud Detection (by: Andrea Dal Pozzolo, Olivier Caelen, Reid A. Johnson and Gianluca Bontempi. Calibrating Probability with Undersampling for Unbalanced Classification. In Symposium on Computational Intelligence and Data Mining (CIDM), IEEE, 2015).  

This dataset is an example of unbalanced data for fraud values are far more scarce than usual.

Import 'pandas' as 'pd' and use “fivethirtyeight” graph style.  

Use 'value_counts' method to count 'Class' target value then use 'plot' to draw a histogram. Then plot another histogram using logarithmic scale.  

Next graph should contain 2 histograms for 'V1' feature: one for fraud values (Class equals 1) and another one for common values (Class equals 0). 

Tune 'density' parameter to bring distribution density. 

'Bin' number shoul be 20, 'alpha' should be 0.5 to make histograms semi-transparent. 

Create a legend with “Class 0” and “Class 1”. The colour of common values histogram should be grey, the colour of fraud values histogram should be red. Make “Class” label for horizontal axis.

```python
df = pd.read_csv('https://datahub.io/machine-learning/creditcard/r/creditcard.csv')
df.head(3)
```

```python
df.Class.replace("'0'", 0, inplace=True)
df.Class.replace("'1'", 1, inplace=True)
```

```python
df.Class.value_counts()
```

```python
plt.style.use('fivethirtyeight')
df.Class.value_counts().plot(kind='bar')
```

```python
plt.style.use('fivethirtyeight')
df.Class.value_counts().plot(kind='bar', logy = True)
```

```python
plt.style.use('fivethirtyeight')
plt.hist(df['V1'][df['Class']==1], alpha=0.5, bins=20, density=True, color='red', label='Class 1')
plt.hist(df['V1'][df['Class']==0], alpha=0.5, bins=20, density=True, color='grey', label='Class 0')
plt.xlabel('Class')
plt.legend()
```
